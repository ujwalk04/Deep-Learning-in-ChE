```python
tf.data.Dataset
data=tf.keras.utils.image_dataset_from_directory(r'C:\Users\mohit\Desktop\assignment\cnn assignment\aug',batch_size=9,image_size=(256,512))
data_iterator=data.as_numpy_iterator()
batch=data_iterator.next()
```
